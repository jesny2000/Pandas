```python
#concatanating two excel records
import pandas as pd

file1 = 'marksheet_1.xlsx'
file2 = 'marksheet_2.xlsx'

sheet1 = pd.read_excel(file1, index_col=0)
sheet2 = pd.read_excel(file2, index_col=0)

# Combine both Excel files
newData = pd.concat([sheet1, sheet2])
print(newData)
```

```python
#Head() and Tail() methods in Pandas
#Return 5 rows
print(newData.head())
print(newData.tail())
```
