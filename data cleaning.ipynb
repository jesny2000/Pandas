```python
import pandas as pd
import numpy as np

d = {'First Score': [100, 90, np.nan, 95],
        'Second Score': [30, 45, 56, np.nan],
        'Third Score': [np.nan, 40, 80, 98]}
df = pd.DataFrame(d)

mv = df.isnull()

print(mv)
```

```python
#Check the gender column and print the rows contain missing data
d = pd.read_csv("employees.csv")
bool_series = pd.isnull(d["Gender"])
missing_gender_data = d[bool_series]
print(missing_gender_data)
```

```python
import pandas as pd
import numpy as np

data = {'Name': ['Amit', 'Sita', np.nan, 'Raj'],
        'Age': [25, np.nan, 22, 28]}

df = pd.DataFrame(data)

# Check for missing values using isna()
print(df.isna())
```

```python
#using notnull()

d = {'First Score': [100, 90, np.nan, 95],
        'Second Score': [30, 45, 56, np.nan],
        'Third Score': [np.nan, 40, 80, 98]}
df = pd.DataFrame(d)

nmv = df.notnull()

print(nmv)
```

```python
#Filter data with non missing values

d = pd.read_csv("employees.csv")

nmg = pd.notnull(d["Team"])

nmgd= d[nmg]

display(nmgd)
```

```python
#replace

import pandas as pd
import numpy as np

data = pd.read_csv("employees.csv")
fd=data.replace(to_replace=np.nan, value=-1001)
fd[1:10]
```

```python
#interpolate()

import pandas as pd
import numpy as np

data = {'Score': [10, np.nan, 30, np.nan, 50]}
df = pd.DataFrame(data)
df['Score'] = df['Score'].interpolate()

print(df)
```

```python
#dropping rows with all null values
dict = {'First Score': [100, np.nan, np.nan, 95],
        'Second Score': [30, np.nan, 45, 56],
        'Third Score': [52, np.nan, 80, 98],
        'Fourth Score': [np.nan, np.nan, np.nan, 65]}
df = pd.DataFrame(dict)
print(df)
df.dropna(how='all')
```

```python
#Drop colums that contain atleast one value

dict = {'First Score': [100, np.nan, np.nan, 95],
        'Second Score': [30, np.nan, np.nan, 56],
        'Third Score': [52, np.nan, 80, 98],
        'Fourth Score': [60, 67, 68, 65]}
df = pd.DataFrame(dict)
print(df)


r=df.dropna(axis=0)  #rows
print(r)
df.dropna(axis=1)   #colums
```

```python
#Dropping Rows with Missing Values in CSV Files

import pandas as pd
d = pd.read_csv("employees.csv")

nd = d.dropna(axis=0, how='any')

print("Old data frame length:", len(d))
print("New data frame length:", len(nd))
print("Rows with at least one missing value:", (len(d) - len(nd)))
```
